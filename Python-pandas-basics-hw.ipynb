# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

data1 = pd.read_csv('movies.csv')
data2 = pd.read_csv('ratings.csv')

merged_inner = pd.merge(left=data1,right=data2, left_on='movieId', right_on='movieId')


a = merged_inner[['title', 'rating']]

a = a [ a.rating == 5.0 ]

a.groupby('title').count().sort_values('rating', ascending=False).head(1)
```
